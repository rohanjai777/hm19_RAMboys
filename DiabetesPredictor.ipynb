```python
var=df.columns[:-1]
var
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import Normalizer
```

```python
def med(lis):
    if(lis[0]<79):
        print("1.Eating sweets can increase your glucose levels")
    elif(lis[0]>79 and lis[0]<122):
        print("1.Your glucose level is fine")
    else:
        print("1.Exercise and control carb intake to lower your glucose levels")
     
    if(lis[1]<62):
        print("2.You have low blood pressure,suggested medicine\n Fludrocortisone, Midodrine")
    elif(lis[1]=>62 and lis[1]<=80):
        print("2.Your blood pressure is fine")
    else:
        print("2.You have high blood pressure\nsuggested medicine  lisinopril (Zestril),benazepril (Lotensin), captopril (Capoten)")
    if(lis[2]<100):
        print("3.You have low insulin,suggested medicine is Sulfonylureas")
    elif(lis[2]=>100 and lis[2]<=130):
        print("3.Your insulin levels are fine")
    else:
        print("3.High insulin levels spotted suggested medicines are as follows:-\n 1.Glucophage \n2.Glucophage XR, \n 3.Fortamet")
```

```python
sc=Normalizer()
y=df.iloc[:,-1]
X=df.drop('Outcome',axis=1)
df=sc.fit_transform(X)
df=pd.DataFrame(df,columns=var)
```

```python
algo= {
    "knn":KNeighborsClassifier(),
    "Decision Tree":DecisionTreeClassifier(),
    "Random Forest":RandomForestClassifier(),
    "SVM" : svm.SVC(),
    "XGBOOST": xgb.XGBClassifier()
    
}
```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
scores={}
for key,value in algo.items():
    val=value
    val.fit(X_train,y_train)
    val.predict(X_test)
    scores[key]=val.score(X_test,y_test)
print(scores)
```
